### 3

```python
AWK_ts=pd.Series(AWK["Adj Close"])
LLY_ts=pd.Series(LLY["Adj Close"])
TDY_ts=pd.Series(TDY["Adj Close"])
VLO_ts=pd.Series(VLO["Adj Close"])
```

```python
AWK_logdiff = np.log(AWK_ts).diff()
LLY_logdiff = np.log(LLY_ts).diff()
TDY_logdiff = np.log(TDY_ts).diff()
VLO_logdiff = np.log(VLO_ts).diff()
```

```python
print(AWK_logdiff.min())
print(AWK_logdiff.max())
print(AWK_logdiff.mean())
print(AWK_logdiff.median())

print(LLY_logdiff.min())
print(LLY_logdiff.max())
print(LLY_logdiff.mean())
print(LLY_logdiff.median())

print(TDY_logdiff.min())
print(TDY_logdiff.max())
print(TDY_logdiff.mean())
print(TDY_logdiff.median())

print(VLO_logdiff.min())
print(VLO_logdiff.max())
print(VLO_logdiff.mean())
print(VLO_logdiff.median())
```
